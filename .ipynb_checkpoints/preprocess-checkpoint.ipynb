# Preprocessing Data

```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import random
random.seed(0) # for validation
```

### Load data
The data set is in matlab's mat file.  
This database consist of a cell array of matrices, each cell is one record part.   
In each matrix each row corresponds to one signal channel:   
1. PPG signal, FS=125Hz; photoplethysmograph from fingertip 
2. ABP signal, FS=125Hz; invasive arterial blood pressure (mmHg) 
3. ECG signal, FS=125Hz; electrocardiogram from channel II

```python
data_path = 'data/'
```

```python
M = scipy.io.loadmat(data_path+'part_1.mat')['p'][0]
```

I used only 84 subjects

```python
M = M[:84]
M.shape
```

### PPG, ECG and ABP signals (Using data from subject 1)

```python
ex_ppg = M[0][0][0:400]
ex_ecg = M[0][2][0:400]
ex_abp = M[0][1][0:400]
```

```python
plt.figure(figsize=(20,5))
plt.plot(ex_ppg)
plt.ylabel('PPG signals')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(ex_ecg)
plt.ylabel('ECG signals')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(ex_abp)
plt.ylabel('ABP signals')
plt.show()
```

## Create Dataset
- PPG and ECG signals are combined as a two dimensional sequence
- ABP is a one dimensional sequence

```python
ppg_ecg = np.array([[M[index][0], M[index][2]] for index in range(M.shape[0])])
ppg_ecg.shape
```

```python
abp = np.array([M[index][1] for index in range(M.shape[0])])
abp.shape
```

**According to the paper:**   
For the final model train/valid set will be combined to train the model, so the actual training dataset is 80%
- Training set: 68%
- Test set: 20%
- Validation set: 12%  

```python
train_per = int(abp.shape[0] * 80 / 100)
test_per = int(abp.shape[0] * 20 / 100)
print('Num training set: ', train_per)
print('Num testing set:  ', test_per)
```

```python
# sort data so, we will be able to train on the larger dataset
sort_index = np.argsort([value.shape[0] for value in abp])

ppg_ecg = np.array([ppg_ecg[index] for index in sort_index])
abp = np.array([abp[index] for index in sort_index])
```

**Split train/test set**

```python
X_main, X_test, y_main, y_test = ppg_ecg[test_per:], ppg_ecg[:test_per], abp[test_per:], abp[:test_per]
```

**Save dataset to files**

```python
np.save(data_path+'x_main.npy', X_main)
np.save(data_path+'y_main.npy', y_main)
np.save(data_path+'x_test.npy', X_test)
np.save(data_path+'y_test.npy', y_test)
```
